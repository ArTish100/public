```python
def GetStory():
  with open("STORY.TXT","w") as F:
    while True:
      T=input("Enter text: ")
      F.write(T+'\n')
      if len(T)==0:
        break
def ViewStory():
  with open("STORY.TXT") as F:
    P=F.read()
    print(P)
def Words():
  with open("STORY.TXT") as F:
    M=F.read()
    for i in M:
      if i!=" ":
        print(i,end="")
      else:
        print()
def AllWords():
  R=0
  with open("STORY.TXT") as F:
    M=F.read()
    for i in M:
      if i!=" ":
        print(i,end="")
      else:
        print()
        R+=1
    print(R)


while True:
  print("G:GetStory V:Viewstory W:Words A:AllWords")
  Choice=input("Enter your choice:")
  if Choice=="G":
    GetStory()
  elif Choice=="V":
    ViewStory()
  elif Choice=="W":
    Words()
  elif Choice=="A":
    AllWords()
  else:
    break
```

```python
def CreateNotes():
  with open("NOTES.TXT","w") as F:
    while True:
      T=input("Enter text: ")
      F.write(T+'\n')
      if len(T)==0:
        break
def CountAVD():
  A=0;B=0;C=0
  with open("NOTES.TXT") as F:
    P=F.read()
    for i in P:
      if i in "aeiou":
        A+=1
      elif i in "qwrtypsdfghjklzxcvbnm":
        B+=1
      elif i in "1234567890":
        C+=1
    print("Vowels:",A)
    print("Consonants:",B)
    print("Numbers:",C)
def ShowNotes():
  with open("NOTES.TXT") as F:
    P=F.read()
    print(P)
def RevText():
  with open("NOTES.TXT") as F:
    while True:
      L=F.readline()
      for i in L:
        if i in "tT":
          N=len(L)
          for i in range(0,N,-1):
            print(i,end="")
while True:
  print("C:CreateNotes AVD:CountAVD")
  Choice=input("Enter your choice:")
  if Choice=="C":
    CreateNotes()
  elif Choice=="AVD":
    CountAVD()
  elif Choice=="S":
    ShowNotes()
  elif Choice=="R":
    RevText()
  else:
    break
```

```python
def Enrol():
  with open("CANDIDATE.DAT","wb") as F:
    while True:
      Candidateno=int(input("Enter enrollment no. of candidate : "))
      Cname=input("Enter name of candidate : ")
      Score=float(input("Enter score of the candidate : "))
      pickle.dump([Candidateno,Cname,Score],F)
      Q=input("More candidates (Y:Yes/N:No) : ")
      if Q=="N":
        break
def GetPass():
  with open("CANDIDATE.DAT","wb") as F:
    L=pickle.load(F)
    for C in L:
      if C[2]>=50:
        print(C)
def ShowAll():
  with open("CANDIDATE.DAT","wb") as F:
    while True:
      try:
        L=pickle.load(F)
        print(L)
      except:
        break
def AverageScore():
  i=0;AVG=0
  with open("CANDIDATE.DAT","wb") as F:
    L=pickle.load(F)
    for C in L:
      AVG+=C[2]
      i+=1
    print("Average score",AVG/i)
while True:
  print("E:Enrol G:GetPass S:ShowAll Avg:AverageScore")
  Choice=input("Enter your choice:")
  if Choice=="E":
    Enrol()
  elif Choice=="G":
    GetPass()
  elif Choice=="S":
    ShowAll()
  elif Choice=="Avg":
    AverageScore()
  else:
    break
```

```python
def Register():
  with open("ACCOUNTS.DAT","wb") as F:
    while True:
      Acno=int(input("Enter account no. : "))
      Name=input("Enter name : ")
      Balance=float(input("Enter balance : "))
      pickle.dump([Acno,Name,Balance],F)
      Q=input("More candidates (Y:Yes/N:No) : ")
      if Q=="N":
        break
def Transact():
  with open("ACCOUNTS.DAT","wb") as F:
    A=input("Enter account no. : ")
    while True:
      C=pickle.load(F)
      for L in C:
        if A==L[0]:
          Ask=input("Would you like to deposit(D) or withdraw(W) ? ")
          if Ask=="D":
            D=int(input("Enter the amount to be deposited : "))
            L[2]+=D
          elif Ask=="W":
            W=int(input("Enter the amount to be withdrawn : "))
            L[2]+=W
def DisplayAll():
  with open("ACCOUNTS.DAT","rb") as F:
    C=pickle.load(F)
    print(C)
def BankBalance():
  H=0
  with open("ACCOUNTS.DAT","rb") as F:
    C=pickle.load(F)
    for L in C:
      H+=L[2]
    print("Total balance :",H)
while True:
  print("R:Register T:Transact A:Display all B:Bank balance")
  Choice=input("Enter your choice:")
  if Choice=="R":
    Register()
  elif Choice=="T":
    Transact()
  elif Choice=="A":
    DisplayAll()
  elif Choice=="B":
    BankBalance()
  else:
    break
```

```python
def Register():
  with open("ACCOUNTS.CSV","w") as cF:
    cW=csv.writer(cF)
    while True:
      Acno=int(input("Enter account no. : "))
      Name=input("Enter name : ")
      Balance=float(input("Enter balance : "))
      M=[Acno,Name,Balance]
      cW.writerow(M)
      Q=input("More candidates (Y:Yes/N:No) : ")
      if Q=="N":
        break
def Transact():
  with open("ACCOUNTS.CSV","w") as cF:
    A=input("Enter account no. : ")
    cW=csv.reader(cF)
    while True:
      for L in cW:
        if A==L[0]:
          Ask=input("Would you like to deposit(D) or withdraw(W) ? ")
          if Ask=="D":
            D=int(input("Enter the amount to be deposited : "))
            L[2]+=D
          elif Ask=="W":
            W=int(input("Enter the amount to be withdrawn : "))
            L[2]+=W
def DisplayAll():
  with open("ACCOUNTS.CSV","r") as cF:
    cR=csv.reader(cF)
    print(cR)
def BankBalance():
  H=0
  with open("ACCOUNTS.CSV","r") as cF:
    cR=csv.reader(cF)
    for L in cR:
      H+=L[2]
    print("Total balance :",H)
while True:
  print("R:Register T:Transact A:Display all B:Bank balance")
  Choice=input("Enter your choice:")
  if Choice=="R":
    Register()
  elif Choice=="T":
    Transact()
  elif Choice=="A":
    DisplayAll()
  elif Choice=="B":
    BankBalance()
  else:
    break
```
